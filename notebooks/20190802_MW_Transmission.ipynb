```python
import os
import sys
sys.path.append('../')

import numpy as np
import matplotlib.pyplot as plt

import src.io as sio
import src.preprocessing as spp
import src.fitting as sft
import ipympl
```

```python
MW_FOLDER = sio.get_folderpath("20190802_MW_Transmission")
```

```python
x = np.genfromtxt(MW_FOLDER + 'C1.csv', delimiter=',', skip_header=3, usecols=(0)) / 1e9
g = np.genfromtxt(MW_FOLDER + 'G.csv', delimiter=',', skip_header=3, usecols=(1))
s = np.genfromtxt(MW_FOLDER + 'S.csv', delimiter=',', skip_header=3, usecols=(1))
```

```python
fig, ax = plt.subplots()
ax.plot(x, s, label="Sample line transmission")
ax.plot(x, g, label="Tip line transmission")

ax.set(xlabel='Frequency (GHz)', ylabel='Transmitted power (dBm)')
D = 2.87
gamma = 28
B1 = 0.05
B2 = 0.1
ax.axvline(D, linestyle="--", color="C7")

ax.axvline(D+gamma*B1, linestyle="--", color="C6")
ax.axvline(D-gamma*B1, linestyle="--", color="C6")

ax.axvline(D+gamma*B2, linestyle="--", color="C5")
ax.axvline(D-gamma*B2, linestyle="--", color="C5")

ax.legend()

fig.tight_layout()
sio.savefig("transmission_sample_tip", folder=MW_FOLDER)
```
